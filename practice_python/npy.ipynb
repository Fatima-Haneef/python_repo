```python
import numpy as np
import matplotlib.pyplot as plt  # for visualization (optional)

def gaussian_2d(size, mean, sigma):
    """
    Generate a 2D Gaussian array.
    
    Parameters:
    - size: tuple of ints (height, width)
    - mean: tuple of floats (mean_x, mean_y)
    - sigma: tuple of floats (sigma_x, sigma_y)
    
    Returns:
    - 2D numpy array with Gaussian values
    """
    x = np.linspace(0, size[1] - 1, size[1])
    y = np.linspace(0, size[0] - 1, size[0])
    x, y = np.meshgrid(x, y)
    
    x0, y0 = mean
    sigma_x, sigma_y = sigma
    
    gauss = (1 / (2 * np.pi * sigma_x * sigma_y)) * np.exp(
        -(((x - x0) ** 2) / (2 * sigma_x ** 2) + ((y - y0) ** 2) / (2 * sigma_y ** 2))
    )
    return gauss

# Example usage
size = (100, 100)
mean = (50, 50)
sigma = (10, 10)

gaussian_array = gaussian_2d(size, mean, sigma)

# Optional: visualize the result
plt.imshow(gaussian_array, cmap='viridis')
plt.colorbar()
plt.title("2D Gaussian Distribution")
plt.show()
```

```python
import numpy as np
def gaussian_2d_simple(size, sigma):
    x = np.linspace(-size // 2, size // 2, size)
    y = np.linspace(-size // 2, size // 2, size)
    x, y = np.meshgrid(x, y)
    d = x**2 + y**2
    gauss = np.exp(-d / (2 * sigma**2))
    return gauss / gauss.sum()  # normalize so sum = 1

gauss = gaussian_2d_simple(100, 10)
print(gauss)
```

```python
import numpy as np
list1 = [1, 2, 3]
list2 = [4, 5, 6]

vector1 = np.array(list1)
vector2 = np.array(list2)
print(str(vector1))
print(str(vector2))

dot_product = vector1.dot(vector2) # dot product of two vectors is a scalars value
print(str(dot_product))  #a1*a2 + a2*b2 + a3*b3
cross_product = np.cross(vector1, vector2) # cross product of two vectors is a vector
print(str(cross_product))  #a2*b3 - a3*b2, a3*b1 - a1*b3, a1*b2 - a2*b1
```
